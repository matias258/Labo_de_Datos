# Ej 13: Empezando con "penguins" escriba una secuencia de operaciones (pipe) que:

- a. Excluya a los pinguinos observados en la isla "Biscoe"
- b. Solo se quede con las variables que estan entre "species" y "body_mass_g" inclusive
- c. Renombre la variable species a "especie-pinguino"
- d. Agrupe los datos por la variable "especie-pinguino"
- e. Calcule el valor medio de las variables que contienen "length", separando por la especie del pinguino y llamando a las colunas como las originales pero agregando "_mean" al final.

```python
# Ahora hagamos una copia con un nuevo dataset  y resumamos todo lo que acabamos de hacer
penguins13bis = penguinsClean.copy()

# a) Quito la isla "Biscoe"
penguins13bis = penguinsClean[penguinsClean["island"] != "Biscoe"]

# b) quito estas 3 columnas y c) renombro "species" y agrupo por "especie de pinguino"
penguins13bis = penguinsClean[[col for col in penguinsClean.columns if col not in ["sex" , "peso_bin" , "body_mass_kg"]]].rename(columns = {"species": "especie_pinguino"})

# d) agrupo por "especie de pinguino", e) calculo length por especie
por_bill_length = penguins13bis.groupby("especie_pinguino")["bill_length_mm"].agg("median")
por_flipper_length = penguins13bis.groupby("especie_pinguino")["flipper_length_mm"].agg("median")

# los agrego como columnas 
penguins13bis["median_bill_length_mm"] =  [por_bill_length[n] for n in penguins13bis["especie_pinguino"]]
penguins13bis["median_flipper_length_mm"] =  [por_flipper_length[n] for n in penguins13bis["especie_pinguino"]]

# y agrego _mean a las columnas con "length" dentro de su nombre (no funca :( )
penguins13bis.columns = [col + "_mean" if "length" in col else col for col in penguins13bis.columns]

penguins13bis
```
